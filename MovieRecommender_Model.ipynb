```python
import numpy as np
import pandas as pd
```

```python
# Step-1: Loading Datasets
data = pd.read_csv("Datasets/ratings.csv")
data.head(10)
```

```python
movie_titles_genre = pd.read_csv("Datasets/movies.csv")
movie_titles_genre.head(10)
```

```python
data = data.merge(movie_titles_genre,on='movieId', how='left')
data.head(10)
```

```python
# Step-2: Feature Engineering
#1.Average Rating
Average_ratings = pd.DataFrame(data.groupby('title')['rating'].mean())
Average_ratings.head(10)
```

```python
#Total Number Of Rating
#_The rating of a movie is proportional to the total number of ratings it has. 
#_Therefore, we will also consider the total ratings cast for each movie.

Average_ratings['Total Ratings'] = pd.DataFrame(data.groupby('title')['rating'].count())
Average_ratings.head(10)
```

```python
# Step-3: Building The Recommender
#Calculating The Correlation

movie_user = data.pivot_table(index='userId',columns='title',values='rating')
```

```python
movie_user.head(10)
```

```python
# Step-4: Now we need to select a movie to test our recommender system.
#1st test => Toy Story (1995).
correlations = movie_user.corrwith(movie_user['Toy Story (1995)'])
correlations.head()
```

```python
# Step-5: Now we will remove all the empty values and merge the total ratings to the correlation table.
recommendation = pd.DataFrame(correlations,columns=['Correlation'])
recommendation.dropna(inplace=True)
```

```python
recommendation = recommendation.join(Average_ratings['Total Ratings'])
recommendation.head()
```

```python
# Step-6: Testing The Recommendation System
#_Let’s filter all the movies with a correlation value to Toy Story (1995) and with at least 100 ratings.

recc = recommendation[recommendation['Total Ratings']>100].sort_values('Correlation',ascending=False).reset_index()
```

```python
#Let’s also merge the movies dataset for verifying the recommendations.
recc = recc.merge(movie_titles_genre,on='title', how='right')
```

```python
#We can see that the top recommendations are pretty good.
recc.head(10)
```
